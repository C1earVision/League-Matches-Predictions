```python
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
df.columns
```

```python
df = df.drop(['blueTeamControlWardsPlaced', 'blueTeamWardsPlaced', 'blueTeamInhibitorsDestroyed', 'blueTeamTurretPlatesDestroyed', 'blueTeamTowersDestroyed', 'blueTeamFirstBlood', 'blueTeamMinionsKilled', 'blueTeamJungleMinions', 'redTeamControlWardsPlaced', 'redTeamWardsPlaced', 'redTeamInhibitorsDestroyed', 'redTeamTowersDestroyed', 'redTeamTurretPlatesDestroyed', 'redTeamMinionsKilled', 'redTeamJungleMinions'], axis=1)
```

```python
df.head()
```

```python
df_new = df.iloc[:, :-1]
df_new.head()
```

```python
df_new = df_new - df_new.shift(-6, axis=1)
df_new = df_new.drop(df_new.iloc[:, 6:], axis=1)
df_new = pd.concat([df_new, df['blueWin']], axis=1)
```

```python
df_new.head()
```

```python
df_new.shape
```

```python
# drop all row that dont make sense ie. should lose but won
col_counter = 0
for index, row in df_new.iterrows():
    for column, value in row.items():
        if value < 0 and df_new['blueWin'][index] == 1:
            col_counter += 1
    if col_counter >= 3:
        df_new = df_new.drop(index)
    col_counter = 0
```

```python
df = df_new
df.tail()
```
